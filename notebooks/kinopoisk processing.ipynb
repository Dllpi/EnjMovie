```python
df['actors'] = df['actors'].replace('Оливия Де Хэвилленд', 'Оливия ДеХэвилленд', regex=True)
```

```python
def extract_actors(actors):
    actors_words = actors.split()
    pairs = [(i,j) for i,j in zip(actors_words, actors_words[1:])]
    return pairs
```

```python
actors_list = list(df['actors'].values)
```

```python
s = [extract_actors(a) for a in actors_list]
```

```python
actors_list = [actor[::2] for actor in s]
```

```python
actors_list_new = []
for actors in actors_list:
    current_actors = []
    for actor in actors:
        current_actors.append(' '.join(list(actor)))
    actors_list_new.append(current_actors)
```

```python
actors_list_new
```

```python
df['actors'] = actors_list_new
```
